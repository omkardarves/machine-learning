# Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# line plot

```python
x=[1,3,5,7,9,11]
y=[11,14,15,19,24,25]
plt.plot(x,y)
```

```python
#To avoid <matplotlib.lines.Line2D at 0x17ef6d54f88> object name
plt.plot(x,y)
plt.show()
```

```python
# sin wave
x=np.arange(0,10,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
area=[600,800,1000,1200,1500]
price=[15,18,25,35,40]
```

```python
plt.plot(area,price)
plt.show()
```

```python
# change the size of plot
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
# adding title
plt.title('Area Vs Price')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
# adding title
plt.title('Area Vs Price')
# adding label
plt.xlabel('Area in sq ft')
plt.ylabel('Price in thousand')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
# adding title
plt.title('Area Vs Price')
# adding label
plt.xlabel('Area in sq ft')
plt.ylabel('Price in thousand')
# change limit/range
plt.xlim((400,2000))
plt.ylim((5,50))
#tick rotation
plt.xticks(rotation=90) # in degrees
plt.yticks(rotation=45)
plt.show()
```

```python
# multiple plots
x=np.arange(1,20)
y1=np.sqrt(x)
y2=np.log(x)
```

```python
plt.plot(x,y1,color='yellow')
plt.plot(x,y2,color='red')
plt.show()
```

```python
plt.plot(x,y1,color='yellow')
plt.plot(x,y2,color='red')
#legend
plt.legend(labels=['sqrt','log'])
plt.show()
```

```python
plt.plot(x,y1,color='yellow')
plt.plot(x,y2,color='red')
plt.plot(x,x,color='blue')
#legend
plt.legend(labels=['sqrt','log','linear'])
plt.show()
```

## Histogram plot

```python
df=pd.read_csv('cars.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#plt.figure(figsize=(7,7))
plt.xlabel('width')
plt.ylabel('frequency')
#how many bars is displayed using bins
plt.hist(df['width'],bins=10)
plt.show()
```

```python
#using pandya
df['width'].hist()
plt.show()
```

```python
df['width'].hist(bins=15)
plt.show()
```

```python
#plt.axvline
plt.hist(df['width'])
width_mean=df['width'].mean()
width_median=df['width'].median()
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='cyan')
plt.show()
```

```python
df['width'].hist()
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='cyan')
plt.show()
```

## Bar plot


```python
#kind bar-vertical and barh-horizontal
#make wise count
df['make'].value_counts()
```

```python
df['make'].value_counts().plot(kind='bar')
plt.xlabel('Make')
plt.ylabel('Count')
plt.show()
```

```python
df['make'].value_counts().plot(kind='barh')
plt.xlabel('Make')
plt.ylabel('Count')
plt.show()
```

# Piestart

```python
#body style wise count
df['body-style'].value_counts()
```

```python
df['body-style'].value_counts().plot(kind='pie')
plt.show()
```

```python
df['body-style'].value_counts().plot.pie()
plt.show()
```

```python
df['body-style'].value_counts().plot(kind='pie',autopct='%1.1f%%') # give value to pie chart in %
plt.show()
```

```python
df['body-style'].value_counts().plot(kind='pie',autopct='%1.1f%%',explode=(0,0,0,0,0.5))
plt.show()
```

```python
df['body-style'].value_counts().plot.pie(autopct='%1.1f%%',explode=(0,0,0,0,0.5))
plt.show()
```
